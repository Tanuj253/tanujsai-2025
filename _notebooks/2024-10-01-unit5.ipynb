popcorn hack 1:
Write a Student class. - Each student has a name - Each student has a score (out of 100) - A student passes if their score is above 70

Check which student passes (write a class header, instance variables, constructors, and methods)

```java
public class Student {
    private String name;
    private int score;
    
    // Default constructor
    public Student() {
        name = "";
        score = 0;   
    }

    // Overloaded constructor
    public Student(String n, int s) {
        name = n;
        setScore(s);  // Use the mutator to ensure score validity
    }

    // Accessor Methods
    public String getName() {
        return name;
    }

    public int getScore(){
        return score;
    }

    // Mutator Methods
    public void setName(String n){
        name = n;
    }

    public void setScore(int s){
        if(s < 0 || s > 100){
            score = 0; // Invalid scores default to 0
        }
        else {
            score = s;
        }
    }

    // Method to check if the student passes
    public boolean pass(){
        return (score >= 70);
    }
}

// Main class to test the Student class
public class Main {
    public static void main(String[] args) {
        // Create some Student objects
        Student student1 = new Student("Alice", 85);
        Student student2 = new Student("Bob", 65);
        Student student3 = new Student("Charlie", 75);
        
        // Print out the students and their pass/fail status
        System.out.println(student1.getName() + " passes: " + student1.pass());
        System.out.println(student2.getName() + " passes: " + student2.pass());
        System.out.println(student3.getName() + " passes: " + student3.pass());
    }
}

Main.main(null);
```

```java
import java.util.ArrayList;
import java.util.List;

public class Villain {
    // Instance variables
    public String name;
    public String evilPlan;
    public List<String> minions;
    public static int villainCount = 0;
    private static int globalMinionCount = 1; // Counter for globally assigned minions

    // Constructor for name, plan, and minions
    public Villain(String name, String evilPlan) {
        this.name = name;
        this.evilPlan = evilPlan;
        this.minions = new ArrayList<>();
        villainCount++;
    }

    // Instance method to add a minion
    public void addMinion(String minion) {
        minions.add(minion);
        System.out.println(minion + " has been added to " + name + "'s army.");
    }

    // Instance method to describe the villain
    public void describeVillain() {
        System.out.println(name + " is planning to: " + evilPlan);
        System.out.println("They have " + minions.size() + " minions.");
    }

    // Static method to get the total count of villains
    public static int getVillainCount() {
        return villainCount;
    }

    // Static method to assign a global minion to all villains
    public static void addGlobalMinion(List<Villain> villains) {
        String newMinion = "Minion " + globalMinionCount;
        globalMinionCount++;

        for (Villain villain : villains) {
            villain.addMinion(newMinion);
        }
    }
}

public class Main {
    public static void main(String[] args) {
        Villain.villainCount = 0;

        // Create new villains
        Villain gru = new Villain("Gru", "steal the moon!");
        Villain vector = new Villain("Vector", "take over the world with magnitude and direction!");

        List<Villain> villains = new ArrayList<>();
        villains.add(gru);
        villains.add(vector);

        System.out.println("=== Adding Minions ===");
        // Create some minions for Gru
        gru.addMinion("Kevin");
        gru.addMinion("Stuart");
        gru.addMinion("Bob");

        // Create some minions for Vector
        vector.addMinion("Henchman 1");

        System.out.println();

        // Describe the villains and their plans
        System.out.println("=== Villain Descriptions ===");
        gru.describeVillain();
        System.out.println();
        vector.describeVillain();
        System.out.println();

        // Add a global minion to all villains
        System.out.println("=== Adding Global Minion to All Villains ===");
        Villain.addGlobalMinion(villains);

        System.out.println();

        // Describe the villains again after the global minion addition
        System.out.println("=== Villain Descriptions After Global Minion ===");
        gru.describeVillain();
        System.out.println();
        vector.describeVillain();
        System.out.println();

        // Get the total count of villains
        System.out.println("=== Total Villain Count ===");
        System.out.println("There are " + Villain.getVillainCount() + " villains in the world.");
    }
}

Main.main(null);
```

```java
public class Minion {
    private String color;      // The color of the minion (e.g., purple or yellow)
    private String name;       // The name of the minion
    private int energyLevel;   // The energy level of the minion
    private String gadget;     // The gadget the minion has
    private String hair;       // The hairstyle of the minion
    private double height;     // The height of the minion in meters

    // Default constructor for NPC minions
    public Minion() {
        this.color = "Yellow";            // Default color
        this.name = "Unnamed Minion";     // Default name
        this.energyLevel = 100;           // Default energy level
        this.gadget = "Banana";           // Default gadget
        this.hair = "None";                // Default hair
        this.height = 1.2;                 // Default height in meters
    }

    // Parameterized constructor to initialize the Minion's attributes
    public Minion(String color, String name, int energyLevel, String gadget, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.gadget = gadget;
        this.hair = hair;
        this.height = height;
    }

    // Getters for the instance variables
    public String getColor() {
        return color;
    }

    public String getName() {
        return name;
    }

    public int getEnergyLevel() {
        return energyLevel;
    }

    public String getGadget() {
        return gadget;
    }

    public String getHair() {
        return hair;
    }

    public double getHeight() {
        return height;
    }

    // Method to print minion details
    public String toString() {
        return "Color: " + color + 
               ", Name: " + name + 
               ", Energy Level: " + energyLevel + 
               ", Gadget: " + gadget + 
               ", Hair: " + hair + 
               ", Height: " + height;
    }
}
```

Hack 5
Gru wants to make sure his workers are not overworked as per OSHA. So, Gru wants you to print out the average energy levels of all his Minions. (Hint: you should use static variables)

```java
// Minion.java
public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private String gadget;
    private String hair;
    private double height;

    // Static variables to track total energy and count of minions
    private static int totalEnergyLevel = 0;
    private static int minionCount = 0;

    // Default constructor
    public Minion() {
        this.color = "Yellow";
        this.name = "Unnamed Minion";
        this.energyLevel = 100;
        this.gadget = "None";
        this.hair = "Bald";
        this.height = 1.0; // Height in meters
        updateEnergyTracking(100); // Update total energy when default minion is created
    }

    // Parameterized constructor
    public Minion(String color, String name, int energyLevel, String gadget, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.gadget = gadget;
        this.hair = hair;
        this.height = height;
        updateEnergyTracking(energyLevel); // Update total energy when parameterized minion is created
    }

    // Method to update the total energy and count of minions
    private void updateEnergyTracking(int energy) {
        totalEnergyLevel += energy;
        minionCount++;
    }

    // Static method to calculate average energy level
    public static double calculateAverageEnergy() {
        return (minionCount > 0) ? (double) totalEnergyLevel / minionCount : 0;
    }

    // toString method to print minion details
    @Override
    public String toString() {
        return "Color: " + color + 
               ", Name: " + name + 
               ", Energy Level: " + energyLevel + 
               ", Gadget: " + gadget + 
               ", Hair: " + hair + 
               ", Height: " + height;
    }
}
```

For 0.90+
Dr. Nefario is trying to assign a recovery time for each minion! Minions who were purple and got cured are very tired, and so are a lot of minions with low energy levels. Create a simple algorithm to calculate how long each minion needs to recover based on their color and energy levels.

```java
// Minion.java
public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private String gadget;
    private String hair;
    private double height;

    // Static variables to track total energy and count of minions
    private static int totalEnergyLevel = 0;
    private static int minionCount = 0;

    // Default constructor
    public Minion() {
        this.color = "Yellow";
        this.name = "Unnamed Minion";
        this.energyLevel = 100;
        this.gadget = "None";
        this.hair = "Bald";
        this.height = 1.0; // Height in meters
        updateEnergyTracking(100); // Update total energy when default minion is created
    }

    // Parameterized constructor
    public Minion(String color, String name, int energyLevel, String gadget, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.gadget = gadget;
        this.hair = hair;
        this.height = height;
        updateEnergyTracking(energyLevel); // Update total energy when parameterized minion is created
    }

    // Method to update the total energy and count of minions
    private void updateEnergyTracking(int energy) {
        totalEnergyLevel += energy;
        minionCount++;
    }

    // Static method to calculate average energy level
    public static double calculateAverageEnergy() {
        return (minionCount > 0) ? (double) totalEnergyLevel / minionCount : 0;
    }

    // Method to calculate recovery time based on color and energy level
    public int calculateRecoveryTime() {
        int recoveryTime = 0;

        // Add time based on color
        if ("Purple".equalsIgnoreCase(color)) {
            recoveryTime += 10; // Base recovery time for purple minions
        } else if ("Yellow".equalsIgnoreCase(color)) {
            recoveryTime += 5; // Base recovery time for yellow minions
        }

        // Add time based on energy level
        if (energyLevel < 50) {
            recoveryTime += 5; // Additional time for low energy
        }

        return recoveryTime;
    }

    // toString method to print minion details
    @Override
    public String toString() {
        return "Color: " + color + 
               ", Name: " + name + 
               ", Energy Level: " + energyLevel + 
               ", Gadget: " + gadget + 
               ", Hair: " + hair + 
               ", Height: " + height;
    }
}
```
