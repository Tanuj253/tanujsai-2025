```python
import java.util.Arrays;

public class PopcornHack {
    public static void main(String[] args) {
        int[][] grassData = {
            {2, 3, 1, 4, 3, 2, 5},
            {1, 2, 3, 1, 4, 2, 3},
            {3, 4, 2, 5, 1, 3, 2}
        };

        int[] totalGrassEaten = sumColumns(grassData);
        System.out.println(Arrays.toString(totalGrassEaten));
    }

    public static int[] sumColumns(int[][] matrix) {
        int rows = matrix.length;
        int cols = matrix[0].length;
        int[] result = new int[cols];

        for (int i = 0; i < cols; i++) {
            for (int j = 0; j < rows; j++) {
                result[i] += matrix[j][i];
            }
        }

        return result;
    }
}
```

```python
public class GrassPasture {
    /** The 2D grid of pasture tastiness values */
    private int[][] pastures;

    /** Constructor initializes the field */
    public GrassPasture(int[][] pastures) {
        this.pastures = pastures;
    }

    /**
     * Returns sum of total tastiness for all values in 2D array
     */
    public int getTotalGrass() {
        int totalSum = 0;
        for (int[] row : pastures) {
            for (int val : row) {
                totalSum += val;
            }
        }
        return totalSum;
    }

    /**
     * Returns max sum of tastiness of a square in the 2D array (square can be 1x1, 2x2, etc.)
     */
    public int maxSquare() {
        int n = pastures.length;
        int m = pastures[0].length;
        int maxSum = Integer.MIN_VALUE;

        // Prefix sum array to optimize square sum calculation
        int[][] prefixSum = new int[n + 1][m + 1];

        // Compute prefix sum
        for (int i = 1; i <= n; i++) {
            for (int j = 1; j <= m; j++) {
                prefixSum[i][j] = pastures[i - 1][j - 1]
                    + prefixSum[i - 1][j]
                    + prefixSum[i][j - 1]
                    - prefixSum[i - 1][j - 1];
            }
        }

        // Iterate over all possible square sizes
        for (int size = 1; size <= Math.min(n, m); size++) {
            for (int i = size; i <= n; i++) {
                for (int j = size; j <= m; j++) {
                    int sum = prefixSum[i][j]
                        - prefixSum[i - size][j]
                        - prefixSum[i][j - size]
                        + prefixSum[i - size][j - size];

                    maxSum = Math.max(maxSum, sum);
                }
            }
        }

        return maxSum;
    }

    /**
     * Returns the maximum tastiness sum subarray in the flattened 2D grid
     */
    public int maxSubarraySum() {
        int[] flattened = flattenGrid();
        return kadane(flattened);
    }

    /** Flattens 2D array into 1D array */
    private int[] flattenGrid() {
        int n = pastures.length, m = pastures[0].length;
        int[] flattened = new int[n * m];
        int index = 0;

        for (int[] row : pastures) {
            for (int val : row) {
                flattened[index++] = val;
            }
        }

        return flattened;
    }

    /** Kadane's Algorithm for Maximum Subarray Sum */
    private int kadane(int[] arr) {
        int maxSum = arr[0], currentSum = arr[0];

        for (int i = 1; i < arr.length; i++) {
            currentSum = Math.max(arr[i], currentSum + arr[i]);
            maxSum = Math.max(maxSum, currentSum);
        }

        return maxSum;
    }

    public static void main(String[] args) {
        int[][] pastures = {
            {-3, 6, -1},
            {2, -1, 5},
            {-9, 4, -1}
        };

        GrassPasture gp = new GrassPasture(pastures);

        System.out.println("Total Tastiness: " + gp.getTotalGrass()); // Expected: -2
        System.out.println("Max Square Sum: " + gp.maxSquare()); // Expected: 9
        System.out.println("Max Subarray Sum: " + gp.maxSubarraySum()); // Expected: 11
    }
}
```
