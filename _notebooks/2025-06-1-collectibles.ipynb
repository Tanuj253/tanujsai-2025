Part 1: Working with Lists

import java.util.ArrayList;
import java.util.List;

public class ListExample {
    public static List<Integer> getEvenNumbers(List<Integer> input) {
        List<Integer> evens = new ArrayList<>();
        for (int i = 0; i < input.size(); i++) {
            int number = input.get(i);
            if (number % 2 == 0) {
                evens.add(number);
            }
        }
        return evens;
    }
}

Uses:
	•	ArrayList
	•	add, get, size
	•	Looping, no streams

Part 2: Working with Sets

import java.util.HashSet;
import java.util.Set;

public class SetExample {
    public static Set<String> getCommonElements(Set<String> set1, Set<String> set2) {
        Set<String> result = new HashSet<>();
        for (String item : set1) {
            if (set2.contains(item)) {
                result.add(item);
            }
        }
        return result;
    }
}

Uses:
	•	HashSet
	•	contains, add
	•	Loop with iteration
	•	Final result has no duplicates (handled by Set)

Part 3: Working with Deques

import java.util.ArrayDeque;
import java.util.Deque;

public class DequeExample {
    public static void simulateLine() {
        Deque<String> line = new ArrayDeque<>();

        // Add 3 customers to the end
        line.addLast("Customer 1");
        line.addLast("Customer 2");
        line.addLast("Customer 3");

        // Add 1 VIP to the front
        line.addFirst("VIP Customer");

        // Remove the customer at the front
        line.removeFirst();

        // Show the front and back of the line
        System.out.println("Front: " + line.peekFirst());
        System.out.println("Back: " + line.peekLast());

        // Print the size of the line
        System.out.println("Line size: " + line.size());
    }
}

Uses:
	•	ArrayDeque
	•	addFirst, addLast, removeFirst, peekFirst, peekLast, size

Challenge Question

Q: You need to store a collection of student IDs where:
	•	Order doesn’t matter
	•	No duplicates allowed
	•	You need to frequently check if an ID exists

A:
Use a HashSet.

Why?
	•	HashSet prevents duplicates automatically.
	•	It provides constant-time (O(1)) lookup for existence checks.
	•	Since order doesn’t matter, there’s no need to use TreeSet or List.
